```python
def search_seven(num):
    """
    input parameters:
    num -> int: integer decimal
    
    returns:
    tst -> bool: true or false boolean depending on whether or not num contains 7.
    """
    str_num = str(num)
    tst = False
    for i in range(len(str_num)):
        if str_num[i] == '7':
            tst = True
            break;
            
    return tst
 
def min_add(num):
    """
    input parameters:
    num -> int: integer decimal
    
    returns:
    glb_min -> int: minimum number of times to add repeated 9's to get a number with
    the digit 7 occuring.
    """
    glb_min = 10
    for i in range(1, len(str(num))+1):
        
        pos_int = int(10**i)-1
        
        sum_res = num 
        
        if search_seven(sum_res):
            glb_min = 0
            break;
        else:
            min_num = 0
            while not search_seven(sum_res):
                sum_res += pos_int
                min_num += 1
            
        if min_num < glb_min:
            glb_min = min_num
    
    return glb_min
    
```
